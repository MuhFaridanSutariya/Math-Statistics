```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import Math
```

```python
#v = vector 
#sk = skalar 
v = np.array([1,2])
sk = [-0.5,1]

for i in sk:
    sk_v = i*v
    plt.plot([0, sk_v[0]],[0, sk_v[1]], 'o-', linewidth=3, label='skalar %g' %i)
    
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
#dikarenakan perkalian skalar 0.5 dengan vector menghasilkan nilai mines maka garisnya akan turun dibawah 0
#setiap perkalian pertama vector dimulai ke kanan lalu ke atas untuk 2 dimensi
```

```python
#v = vector 
#sk = skalar 
v = np.array([1,2])
sk = [-0.5,1, 1.5]

for i in sk:
    sk_v = i*v
    plt.plot([0, sk_v[0]],[0, sk_v[1]], 'o-', linewidth=3, label='skalar %g' %i)
    
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
#v = vector 
#sk = skalar 
v = np.array([1,2])
sk = [-0.5,1, 1.5, 3]

for i in sk:
    sk_v = i*v
    plt.plot([0, sk_v[0]],[0, sk_v[1]], 'o-', linewidth=3, label='skalar %g' %i)
    
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

# penjumlahan dua vector

```python
v1 = np.array([1,2])
v2 = np.array([2,1])

v3a = v1+v2
v3b = np.add(v1,v2)

print(v3a, v3b)
```

```python
plt.plot([0, v1[0]], [0, v1[1]], label='v1', linewidth=2)
plt.plot([0, v2[0]], [0, v2[1]], label='v2', linewidth=2)
plt.plot([0, v3a[0]], [0, v3a[1]], label='v3a', linewidth=2)

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot([0, v1[0]], [0, v1[1]], label='v1', linewidth=2)
plt.plot([0, v2[0]]+v1[0], [0, v2[1]]+v1[1], label='v2', linewidth=2)
plt.plot([0, v3a[0]], [0, v3a[1]], label='v3a', linewidth=2)

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
p1 = sym.latex(sym.sympify(v1))
p2 = sym.latex(sym.sympify(v2))
p3 = sym.latex(sym.sympify(v3a))

display(Math('%s+%s=%s' %(p1,p2,p3)))
```
